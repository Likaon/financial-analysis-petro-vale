Para analisar qual das ações tem mostrado maior crescimento constante, podemos usar algumas ferramentas de análise técnica e de desempenho de ações. 

Taxa de Crescimento Composta Anual (CAGR): Mede o crescimento médio anual de uma ação, considerando um intervalo de tempo. Ela pode apresentar uma visão clara de qual ação tem gerado mais valor durante o período analisado.

Comparação de Médias Móveis: A média móvel de 50 dias (curto prazo) e a média móvel de 200 dias (longo prazo) podem ser usadas para entender a tendência de crescimento e a consistência das ações.

1. Taxa de Crescimento Composta Anual (CAGR)
A fórmula para o CAGR é:

![Texto alternativo](Calc.png)

Onde:

Valor Final é o valor mais recente da ação (preço de fechamento).
Valor Inicial é o valor de 2 anos atrás.
Número de anos é 2, pois estamos analisando os últimos dois anos.

2. Comparação das Médias Móveis
Como já mostramos no código anterior, as médias móveis de 50 dias e 200 dias podem ser usadas para observar o comportamento de curto e longo prazo. Se a média de 50 dias estiver acima da de 200 dias, isso geralmente indica uma tendência de alta.

```python
# Calculando o CAGR para PETR4.SA
initial_price_petr4 = data_petr4['Close'].iloc[0]  # Preço inicial (2 anos atrás)
final_price_petr4 = data_petr4['Close'].iloc[-1]  # Preço final (hoje)
years_petr4 = 2  # Intervalo de 2 anos
cagr_petr4 = (final_price_petr4 / initial_price_petr4) ** (1 / years_petr4) - 1

# Calculando o CAGR para VALE3.SA
initial_price_vale3 = data_vale3['Close'].iloc[0]
final_price_vale3 = data_vale3['Close'].iloc[-1]
years_vale3 = 2
cagr_vale3 = (final_price_vale3 / initial_price_vale3) ** (1 / years_vale3) - 1

# Plotando o gráfico comparativo das duas ações com as médias móveis
plt.figure(figsize=(14, 8))

# Plotando os dados de PETR4.SA
plt.plot(data_petr4['Close'], label="PETR4.SA Fechamento", color="blue", linewidth=2)
plt.plot(data_petr4['MA50'], label="Média Móvel 50 dias PETR4.SA", color="orange", linestyle="--")
plt.plot(data_petr4['MA200'], label="Média Móvel 200 dias PETR4.SA", color="green", linestyle="--")

# Plotando os dados de VALE3.SA
plt.plot(data_vale3['Close'], label="VALE3.SA Fechamento", color="red", linewidth=2)
plt.plot(data_vale3['MA50'], label="Média Móvel 50 dias VALE3.SA", color="purple", linestyle="--")
plt.plot(data_vale3['MA200'], label="Média Móvel 200 dias VALE3.SA", color="brown", linestyle="--")

# Adicionando título e rótulos aos eixos
plt.title('Comparação de Preços de Ações: PETR4.SA vs VALE3.SA (últimos 2 anos)', fontsize=16)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Preço de Fechamento (R$)', fontsize=12)

# Adicionando a legenda
plt.legend(loc='upper left')

# Exibir o gráfico com grid para facilitar a leitura
plt.grid(True)

# Exibir o gráfico
plt.tight_layout()
plt.show()

# Exibindo o CAGR de cada ação
print(f"CAGR de PETR4.SA nos últimos 2 anos: {cagr_petr4 * 100:.2f}%")
print(f"CAGR de VALE3.SA nos últimos 2 anos: {cagr_vale3 * 100:.2f}%")
```

O CAGR de cada ação impresso, permite ver qual delas teve maior crescimento anual composto nos últimos dois anos.

Análise:
O CAGR indica qual ação teve maior crescimento percentual anual ao longo do período de dois anos. Uma maior taxa de crescimento significa que o ativo teve um desempenho superior em termos de valorização.

O gráfico permite observar como as duas ações se comportaram ao longo do tempo, identificando períodos de alta ou queda, além de quando as médias móveis indicam reversões ou mudanças de tendência.


<h4>Passo 1:</h4> Identificar os cruzamentos (quando a média de 50 dias cruza a de 200 dias).

```python
# Baixando os dados históricos da PETR4 e VALE3
acoes = ['PETR4.SA', 'VALE3.SA']
dados = {acao: yf.download(acao, start='2023-01-01', end='2024-01-01') for acao in acoes}

# Exibindo as primeiras linhas dos dados de cada ação
for acao in dados:
    print(f"\n{acao} - Primeiras linhas:")
    print(dados[acao].head())
```

<h4>Passo 2:</h4> Calcular as Médias Móveis de 50 e 200 Dias para ambas as ações.

```python
# Calculando as médias móveis de 50 e 200 dias para ambas as ações
for acao in dados:
    dados[acao]['MA50'] = dados[acao]['Close'].rolling(window=50).mean()
    dados[acao]['MA200'] = dados[acao]['Close'].rolling(window=200).mean()

# Exibindo as primeiras linhas para verificar as médias móveis
for acao in dados:
    print(f"\n{acao} - Médias Móveis:")
    print(dados[acao][['Close', 'MA50', 'MA200']].head())
```

<h4>Passo 3:</h4> Identificar os Cruzamentos das médias móveis.

```python
# Identificar cruzamentos para ambas as ações
for acao in dados:
    dados[acao]['Crossover'] = dados[acao]['MA50'] > dados[acao]['MA200']  # Verifica quando a MA50 é maior que a MA200
    dados[acao]['Crossover_shifted'] = dados[acao]['Crossover'].shift(1)  # Desloca a série para verificar o cruzamento
    
    # Cruzamentos de alta (MA50 cruza MA200 para cima)
    crossover_up = dados[acao][(dados[acao]['Crossover'] == True) & (dados[acao]['Crossover_shifted'] == False)]
    
    # Cruzamentos de baixa (MA50 cruza MA200 para baixo)
    crossover_down = dados[acao][(dados[acao]['Crossover'] == False) & (dados[acao]['Crossover_shifted'] == True)]
    
    # Exibindo os cruzamentos encontrados
    print(f"\n{acao} - Cruzamentos de Alta:")
    print(crossover_up[['Close', 'MA50', 'MA200']])
    
    print(f"\n{acao} - Cruzamentos de Baixa:")
    print(crossover_down[['Close', 'MA50', 'MA200']])
```

<h4>Passo 4:</h4> Gráfico Comparativo, com plot dos preços de fechamento e das medias móveis para ambas as ações.

```python
import matplotlib.pyplot as plt

# Criando o gráfico comparativo
plt.figure(figsize=(14,8))

# Plotando o preço de fechamento e as médias móveis para PETR4 e VALE3
for acao in dados:
    plt.plot(dados[acao].index, dados[acao]['Close'], label=f'{acao} - Preço de Fechamento')
    plt.plot(dados[acao].index, dados[acao]['MA50'], label=f'{acao} - MA50')
    plt.plot(dados[acao].index, dados[acao]['MA200'], label=f'{acao} - MA200')

    # Identificando os cruzamentos de alta (setas verdes) e de baixa (setas vermelhas)
    crossover_up = dados[acao][(dados[acao]['Crossover'] == True) & (dados[acao]['Crossover_shifted'] == False)]
    crossover_down = dados[acao][(dados[acao]['Crossover'] == False) & (dados[acao]['Crossover_shifted'] == True)]
    
    # Plotando os cruzamentos de alta
    plt.scatter(crossover_up.index, crossover_up['Close'], marker='^', color='green', label=f'{acao} - Cruzamento de Alta', alpha=0.7)
    
    # Plotando os cruzamentos de baixa
    plt.scatter(crossover_down.index, crossover_down['Close'], marker='v', color='red', label=f'{acao} - Cruzamento de Baixa', alpha=0.7)

# Configurações do gráfico
plt.title('Cruzamentos de Médias Móveis (50 e 200 dias) - PETR4 vs VALE3')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.legend(loc='best')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

Cruzamentos de Alta: Quando a média móvel de 50 dias (curto prazo) cruza para cima a média de 200 dias (longo prazo), indicando uma tendência de alta.

Cruzamentos de Baixa: Quando a média móvel de 50 dias cruza para baixo a média de 200 dias, indicando uma tendência de baixa.

Vou abordar um estudo sobre RSI e analisar os cruzamentos das médias móveis para identificar possíveis pontos de reversão para PETR4 e VALE3.

<h4>Passo 1:</h4> Calcular o Índice de Força Relativa (RSI)
O RSI é um indicador técnico que mede a velocidade e a mudança dos movimentos de preço, sendo um ótimo indicador para detectar condições de sobrecompra (acima de 70) ou sobrevenda (abaixo de 30).

```python
# Função para calcular o RSI
def calcular_rsi(df, period=14):
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()

    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Calculando o RSI para ambas as ações (PETR4 e VALE3)
for acao in dados:
    dados[acao]['RSI'] = calcular_rsi(dados[acao])

# Exibindo as primeiras linhas do RSI calculado
for acao in dados:
    print(f"\n{acao} - RSI:")
    print(dados[acao][['Close', 'RSI']].tail())
```

Cruzamentos de médias móveis: Considerando a oportunidade de compra quando a MA50 cruza acima da MA200 e a oportunidade de venda quando o oposto acontece.

RSI: Se o RSI estiver acima de 70, isso pode indicar uma possível condição de sobrecompra (momento de venda). Se o RSI estiver abaixo de 30, isso pode indicar uma possível condição de sobrevenda (momento de compra).

```python
# Analisando os cruzamentos e RSI
for acao in dados:
    # Cruzamentos de alta (MA50 cruza MA200 para cima) e cruza para baixo (oposto)
    crossover_up = dados[acao][(dados[acao]['Crossover'] == True) & (dados[acao]['Crossover_shifted'] == False)]
    crossover_down = dados[acao][(dados[acao]['Crossover'] == False) & (dados[acao]['Crossover_shifted'] == True)]
    
    # Padrões de Reversão pelo RSI
    rsi_compra = dados[acao][dados[acao]['RSI'] < 30]  # Quando o RSI está abaixo de 30
    rsi_venda = dados[acao][dados[acao]['RSI'] > 70]   # Quando o RSI está acima de 70
    
    # Exibindo oportunidades de compra e venda (com base no RSI)
    print(f"\n{acao} - Oportunidades de Compra (RSI < 30):")
    print(rsi_compra[['Close', 'RSI']])
    
    print(f"\n{acao} - Oportunidades de Venda (RSI > 70):")
    print(rsi_venda[['Close', 'RSI']])

    # Exibindo cruzamentos de médias móveis
    print(f"\n{acao} - Cruzamentos de Alta:")
    print(crossover_up[['Close', 'MA50', 'MA200']])
    
    print(f"\n{acao} - Cruzamentos de Baixa:")
    print(crossover_down[['Close', 'MA50', 'MA200']])
```

```python
plt.figure(figsize=(14,8))

# Plotando o preço de fechamento e as médias móveis
for acao in dados:
    plt.plot(dados[acao].index, dados[acao]['Close'], label=f'{acao} - Preço de Fechamento')
    plt.plot(dados[acao].index, dados[acao]['MA50'], label=f'{acao} - MA50')
    plt.plot(dados[acao].index, dados[acao]['MA200'], label=f'{acao} - MA200')

    # Destacando os pontos de compra (RSI < 30) e venda (RSI > 70)
    plt.scatter(rsi_compra.index, rsi_compra['Close'], marker='^', color='green', label=f'{acao} - Compra (RSI < 30)', alpha=0.7)
    plt.scatter(rsi_venda.index, rsi_venda['Close'], marker='v', color='red', label=f'{acao} - Venda (RSI > 70)', alpha=0.7)

# Configurações do gráfico
plt.title('Oportunidades de Compra e Venda - PETR4 vs VALE3')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.legend(loc='best')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```
